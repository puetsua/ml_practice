```python
import csv
import numpy as np
import sklearn.datasets as skdata
import matplotlib.pyplot as plt
import matplotlib.colors as mplColors
%matplotlib inline
```

```python
data, labels = skdata.make_moons(200, True, 0.08, 1991)
npdata = np.array(data)
npdata = np.c_[ npdata, np.array(labels) ]

colors = ['red','green']

plt.scatter(npdata[:,0], npdata[:,1], c=labels, cmap=mplColors.ListedColormap(colors))
plt.show()
```

```python
with open('twomoon.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(npdata)
```
